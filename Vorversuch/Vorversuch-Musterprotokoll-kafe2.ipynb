**Aufgabe 2.2**, **Aufgabe 2.3**, **Aufgabe 3.1**, **Aufgabe 3.2**

 * Hier finden Sie Implementierungen der folgenden Modelle:

$$
\begin{split}
&\varphi(x; x_{0}, T, \phi_{0}) = x_{0}\cos\left(\frac{2\pi}{T}x+\phi\right)\\
&\varphi(x; x_{0}, g, \phi_{0}; \ell) = x_{0}\cos\left(\sqrt{\frac{g}{\ell}}x+\phi\right)\\
&\varphi(x; x_{0}, g, \phi_{0}; M, s, \Theta) = x_{0}\cos\left(\sqrt{\frac{M\,s\,g}{\Theta}}x+\phi\right)\\
&\varphi(x; x_{0}, g, \phi_{0}, \tau; M, s, \Theta) = x_{0}e^{-x/\tau}\cos\left(\sqrt{\frac{M\,s\,g}{\Theta}-\frac{1}{\tau^{2}}}x+\phi\right)\\
\end{split}
$$

 * Wir demonstireren die folgenden Schritte: 
 
   * Definition eines Modelles, als *python* Funktion; 

   * Anpassung des Modells an die Daten;
   
   * Ausgabe der Ergebnisse der Anpassung;
   
   * Darstellung der Daten und der Modellfunktionen;
   
   * Darstellung zusätzlicher Kontouren, die zum weiteren Verständnis der Unsicherheiten auf die Modellparameter nützlich sein können.

```python
# Import required functions and classes
import numpy as np
from kafe2 import Fit, Plot, ContoursProfiler

# Fit model Aufgabe 2.3
def harmonic(x, x0=0.75, T=1.59, phi0=0.):
    return x0*np.cos(2*np.pi/T*x+phi0)
# Fit model Aufgabe 2.3
def g_harmonic(x, x0=0.75, g=9.8, phi0=0):
      return x0*np.cos(np.sqrt(g/0.6285)*x+phi0)
# Fit model Aufgabe 3.1
def g_harmonic_new(x, x0=0.75, g=9.8, phi0=0):
      return x0*np.cos(np.sqrt(0.789*0.473*g/0.23523)*x+phi0)
# Fit model Aufgabe 3.2
def g_damped(x, tau=240., x0=0.75, g=9.8, phi0=0):
      return x0*np.exp(-x/tau)*np.cos(np.sqrt(0.789*0.473*g/0.23523-1./tau**2)*x+phi0)
    
# Declare and do the fit (below the model g_damped is passed) 
fit = Fit(xy_data, model_function=g_damped)
fit.do_fit()
# Return a reasonably clean report
fit.report(show_data=False, show_model=False)

# Plot the result
plot = Plot(fit_objects=fit)
plot.x_label = '$t\,(\mathrm{s})$'
plot.y_label = '$a\,(\mathrm{m/s^{2}})$'
plot.plot()  

# Create contour plots
cpf = ContoursProfiler(fit)
cpf.plot_profiles_contours_matrix()

# Show plot in notebook
plot.show()
```
