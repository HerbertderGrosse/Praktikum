verwenden. Dieser Wert wurde aus der [Global Gravtiy Database des Bureau Gravimetrique International (BGI)](https://ggos.org/item/bgi/) für die Stadt Mannheim (bei $49,49^{\circ}$ nördlicher Breite und $8,53^{\circ}$ westlicher Länge auf einer Referenzhöhe von $101\,\mathrm{m}$) ausgelesen.

## Aufgabe 3: Schrittweise Verbesserungen des Modells

Eine offensichtliche Unzulänglichkeit des vorherigen Modells besteht in der Vernachlässigung der endlichen Ausdehnung des Pendels. Wenn Sie ihr Modell entsprechend erweitern, nimmt die Formel zur Bestimmung von $g$ die folgende Form an: 

$$
\begin{equation*}
g = \frac{4\,\pi^{2}}{T^{2}}\frac{\Theta}{m\,\ell},
\end{equation*}
$$

wobei $\Theta$ und $m$ dem Trägheitsmoment und der Masse der gesamten Pendelkonstruktion (einschließlich aller Montageteile und Smartphone!) und $\ell$ dem Abstand des Schwerpunkts der Konstruktion von der Aufhängung des Pendels entsprechen.

### Aufgabe 3.1: Zweite Verbesserung des Modells

### Aufgabe 3.2: Direkte Bestimmung von $g$

 * Beachten Sie, dass das der so bestimmte Wert von $\Delta g$ nur die Unsicherheiten der Datenpunkte widerspiegelt. Diskutieren Sie, wie Sie den Einfluss der Unsicherheiten auf $\Theta$, $m$, und $\ell$ auf $\Delta g$ abschätzen können.

### Aufgabe 3.3: Übergang zur [gedämpften Schwingung](https://de.wikipedia.org/wiki/Schwingung#Linear_ged%C3%A4mpfte_Schwingung) 

Das Pendel erfährt in seiner Bewegung zusätzlich eine Dämpfung. Unterlegen Sie der Messung das Modell einer linearer gedämpften Schwingung zugrunde verändert sich die Formel zur Bestimmung von $g$ wie folgt: 

$$
\begin{equation*}
g = \left(\frac{4\,\pi^{2}}{T_{0}^{2}}+\frac{1}{\tau^{2}}\right)\frac{\Theta}{m\,\ell},
\end{equation*}
$$

wobei $\tau$ einer Abklingzeit der Schwingungsamplitude in Sekunden entspricht. Wie Sie sehen handelt es sich um eine Korrektur, die die Abschätzung von $g$ zu größeren Werten hin beeinflusst.

  * Wie könnten Sie die Hypothese, dass das zugrundeliegende Modell die Daten beschreiben kann, besser testen? 
  * Wie groß ist der Einfluss, den dieser Effekt auf die Messung von $g$ hat?

## Bonusaufgabe: Vom Messen zur Kunst

Ein bewusst ausgewähltes Modell zur Behandlung der Unsicherheiten, die in eine Parameterschätzung eingehen, ist Bestansteil eines guten statistischen Modells. Die Diskussion über die Berücksichtigung der Unsicherheiten der Modellparameter $\Theta$, $m$ und $\ell$, bei der Bestimmung von $g$ und $\Delta g$ (**Aufgabe 3.2**) wirft eine weitere Frage auf, die wir im Rahmen dieses Vorversuchs bisher offen gelassen haben: Sind die Unsicherheiten auf diese zusätzlichen Parameter der Messung korreliert; wenn ja, wie? Sie können die folgenden Bonusaufgaben bearbiten, um dieser Frage weiter nachzugehen. Die Bearbeitung ist jedoch nicht verpflichtend.

### Bonusaufgabe 1: Korrelierte Unsicherheiten

Jede Variation eines der drei Parameter $\Theta$, $m$ oder $\ell$ in Aufgabe 3.2 hat einen nicht-trivialen Einfluss nicht nur auf $g$, sondern auch auf die jeweils anderen Parameter. Durch naive, quadratische Addition von $\Delta \Theta$, $\Delta m$, und $\Delta \ell$ unterlegen Sie (vielleicht unterbewusst) die Annahme, das alle drei Variationen unabhängig sind. Diese Annahme ist auf jeden Fall falsch. Ein anderes Modell, dass Sie anwenden könnten, wäre zwei oder alle Parameter vollständig zu korrelieren. Was bedeutet diese Annahme für die Variation der Parameter? Denken, Sie dass diese Annahme korrekt ist? Machen Sie einen Vorschlag zur Lösung dieses Problems.
